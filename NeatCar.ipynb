```python
import pygame
import neat
import numpy as np
import random
import os
from math import sin,cos,pi,sqrt
```

```python
L = np.array([100,100,50])
v = np.array([0.75,0.75,1])
pygame.init()
font = pygame.font.Font('freesansbold.ttf',32)
win = pygame.display.set_mode((600,600))
win.fill ([255,255,255])
pygame.display.set_caption("Robotic Arm")
```

```python
class Car:
    def __init__(self):
        self.x = random.randint(150,451)
        self.y = random.randint(150,451)
        self.v = 2
    def Draw(self):
        pygame.draw.circle(win,[250,3,0],[self.x,self.y],10,4)
    def up(self):
        self.y -= self.v
    def down(self):
        self.y += self.v
    def right(self):
        self.x += self.v
    def left(self):
        self.x -= self.v
    
```

```python
def Object():
    r = random.randint(0,251)
    o = random.random()*2*pi
    x = r*cos(o) +300
    y = r*sin(o) +300
    return np.array([x,y])
```

```python
def DisplayFitness(value):
    score = font.render('fitness = '+str(value),True,[30,30,250])
    win.blit(score,[10,10])
```

```python
def eval_genomes(genomes, config):
    nets = []
    cars = []
    ball = Object()
    for id, g in genomes:
        net = neat.nn.FeedForwardNetwork.create(g, config)
        nets.append(net)
        g.fitness = 0
        car = Car()
        cars.append(car)
        
        
    running = True
    counter = 0
    while running:
        pygame.time.delay(3)
        pygame.display.update()
        for event in pygame.event.get():
             if event.type  == pygame.QUIT:
                running = False
                pygame.quit()
        
        win.fill ([255,255,255])        
        for i, car in enumerate(cars):
            d1=ball[0]-cars[i].x
            d2=ball[1]-cars[i].y
            x1 = abs(ball[0]-cars[i].x)
            y1 = abs(ball[1]-cars[i].y)
            output = nets[i].activate((d1,d2))
            if output[0] > 0.5:
                cars[i].up()
            if output[0] <- 0.5:
                cars[i].down()
            if output[1] > 0.5:
                cars[i].right()
            if output[1] <- 0.5:
                cars[i].left()
            cars[i].Draw()                
            pygame.draw.circle(win,[30,30,250],[int(ball[0]),int(ball[1])],10,4) 
            x2 = abs(ball[0]-cars[i].x)
            y2 = abs(ball[1]-cars[i].y)
            if (x2<x1 and y2<y1) or (x2<1 and y2<1 and x2 == x1 and y2==y1):               
                genomes[i][1].fitness += 10
            else:
                genomes[i][1].fitness -= 10
        counter = counter+ 1
        if counter == 300:            
            running = False

        
        
    
```

```python
def run(config_file):
 
    config = neat.config.Config(neat.DefaultGenome, neat.DefaultReproduction,
                         neat.DefaultSpeciesSet, neat.DefaultStagnation,
                         config_file)
    p = neat.Population(config)

    p.add_reporter(neat.StdOutReporter(True))
    stats = neat.StatisticsReporter()
    p.add_reporter(stats)
    winner = p.run(eval_genomes, 1000000)
```
